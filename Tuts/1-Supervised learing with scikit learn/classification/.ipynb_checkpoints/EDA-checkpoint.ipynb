# Exploratory data analysis (EDA)

```python
x=iris.data
```

```python
y=iris.target
```

```python
x,y
```

```python
df=pd.DataFrame(x,columns=iris.feature_names)
df.head()
```

```python
pd.plotting.scatter_matrix(df,c=y,figsize=[8,8],marker='D',s=150)
plt.show()
```
